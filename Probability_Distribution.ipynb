```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

```python
lambda_value = 4  # Average number of events (e.g., customers per hour)

# Define the range of values (k) to evaluate
k_values = np.arange(0, 15)  # Possible number of events (0 to 14)
```

```python
poisson_probs = poisson.pmf(k_values, lambda_value)
```

```python
print("Poisson Probabilities:")
for k, prob in zip(k_values, poisson_probs):
    print(f"P(X = {k}) = {prob:.4f}")
```

```python
plt.figure(figsize=(8, 5))
plt.bar(k_values, poisson_probs, color='skyblue', alpha=0.7, edgecolor='black')
plt.title("Poisson Distribution (λ = 4)", fontsize=16)
plt.xlabel("Number of Events (k)", fontsize=12)
plt.ylabel("Probability", fontsize=12)
plt.xticks(k_values)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

how to simulate random events using the Poisson distribution and visualize the results as a histogram.

The line simulated_events = poisson.rvs(mu=lambda_value, size=1000) generates 1000 random events based on a Poisson distribution with a mean 𝜆 = 4.

Each value in the simulated_events array represents the number of events that occurred in one simulated interval (e.g., customers arriving in an hour).

The poisson.rvs function comes from the scipy.stats module and is used for random sampling from a Poisson distribution.

The plt.hist function is used to create a histogram of these simulated events. The bins=range(0, 15) parameter divides the histogram into discrete bins from 0 to 14, which represent the possible counts of events. The density=True option ensures that the histogram displays the relative frequency (probabilities) rather than raw counts, allowing it to be comparable to a theoretical Poisson distribution.

The visual appearance of the histogram is enhanced with additional styling:

color='lightgreen' fills the bars with a light green color.
edgecolor='black' adds a black outline to each bar.
alpha=0.7 sets the transparency of the bars.

Finally, the plot is titled "Histogram of Simulated Events (λ = 4)" to indicate what it represents. The x-axis (Number of Events (k)) and y-axis (Relative Frequency) are labeled, and a grid is added for better readability. This histogram gives a clear picture of how the simulated data aligns with the Poisson distribution's expected behavior.

```python
# Simulate random events using the Poisson distribution
simulated_events = poisson.rvs(mu=lambda_value, size=1000)
```

```python
simulated_events
```

```python
plt.figure(figsize=(8, 5))
plt.hist(simulated_events, bins=range(0, 15), density=True, color='lightgreen', edgecolor='black', alpha=0.7)
plt.title("Histogram of Simulated Events (λ = 4)", fontsize=16)
plt.xlabel("Number of Events (k)", fontsize=12)
plt.ylabel("Relative Frequency", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Bernoulli Distribution**

image.png

image.png

image.png

image.png

image.png

demonstrates the Bernoulli distribution by simulating a random experiment with a given probability of success(p) and visualizing the outcomes.

In this example,
p=0.6, meaning there is a 60% chance of success (represented by 1) and a 40% chance of failure (represented by 0).

The bernoulli.rvs(p, size=num_trials) function from the scipy.stats library is used to generate 1000 trials (random outcomes) based on the specified probability. These outcomes are stored in the data array, which contains only 0s and 1s.

The script then calculates the relative frequencies of success (1) and failure (0) using np.unique(data, return_counts=True) to count occurrences of each outcome, and divides these counts by the total number of trials (num_trials).

The frequencies provide an empirical approximation of the probabilities, which should be close to p for success and 1−p for failure.

Finally, a bar chart is plotted to visually represent the distribution of outcomes. The chart has two bars: one for failure (0) and another for success (1). The heights of the bars correspond to their respective probabilities. Additional formatting (e.g., gridlines, colors, and labels) enhances the clarity of the visualization, providing a clear understanding of how the Bernoulli distribution behaves for the given probability.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
```

```python
p = 0.6  # Example: 60% chance of success

# Generate Bernoulli random variables (0s and 1s)
num_trials = 1000
data = bernoulli.rvs(p, size=num_trials)

# Calculate the relative frequencies of 0s and 1s
unique, counts = np.unique(data, return_counts=True)
frequencies = counts / num_trials
```

```python
print(f"Probability of Success (1): {frequencies[1]:.2f}")
print(f"Probability of Failure (0): {frequencies[0]:.2f}")
```

```python
plt.figure(figsize=(6, 4))
plt.bar([0, 1], frequencies, color='lightblue', alpha=0.7, edgecolor='black')
plt.xticks([0, 1], ['Failure (0)', 'Success (1)'])
plt.ylabel("Relative Frequency")
plt.title(f"Bernoulli Distribution (p = {p})", fontsize=14)
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Binomial distribution**

image.png

image.png

image.png

image.png

image.png

This code demonstrates the Binomial distribution, which models the number of successes in a fixed number of independent trials, each with the same probability of success.

In this example, we set the number of trials (n) to 10 and the probability of success (p) to 0.5.

First, the code calculates the theoretical probabilities of achieving k successes (where k ranges from 0 to n) using the binom.pmf() function from the scipy.stats library. These probabilities are stored and displayed alongside the corresponding k values.

Next, a bar chart is created to visualize the theoretical distribution, where the x-axis represents the number of successes (k), and the y-axis represents their probabilities. This chart provides an intuitive view of how the probability varies with the number of successes in 10 trials.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
```

```python
n = 10  # Number of trials
p = 0.5  # Probability of success in each trial

# Generate possible outcomes (0 to n successes)
k_values = np.arange(0, n + 1)

# Calculate Binomial probabilities for each value of k
binomial_probs = binom.pmf(k_values, n, p)
```

```python
print("Binomial Probabilities:")
for k, prob in zip(k_values, binomial_probs):
    print(f"P(X = {k}) = {prob:.4f}")
```

```python
plt.figure(figsize=(8, 5))
plt.bar(k_values, binomial_probs, color='skyblue', alpha=0.7, edgecolor='black')
plt.title(f"Binomial Distribution (n = {n}, p = {p})", fontsize=16)
plt.xlabel("Number of Successes (k)", fontsize=12)
plt.ylabel("Probability", fontsize=12)
plt.xticks(k_values)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

To simulate real-world outcomes, the code uses binom.rvs() to generate 1000 random samples representing the number of successes in 10 trials. A histogram of these simulated outcomes is then plotted, showing the relative frequencies of each k. The histogram closely matches the theoretical bar chart, illustrating that the simulation aligns with the expected Binomial distribution. This combination of theoretical and simulated results helps in understanding both the formula and its application.

```python
simulated_data = binom.rvs(n, p, size=1000)  # Generate 1000 trials

# Plot histogram of simulated outcomes
plt.figure(figsize=(8, 5))
plt.hist(simulated_data, bins=np.arange(-0.5, n + 1.5, 1), density=True, color='lightgreen', edgecolor='black', alpha=0.7)
plt.title(f"Histogram of Simulated Binomial Outcomes (n = {n}, p = {p})", fontsize=16)
plt.xlabel("Number of Successes (k)", fontsize=12)
plt.ylabel("Relative Frequency", fontsize=12)
plt.xticks(k_values)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Geometric Distribution**

image.png

image.png

image.png

image.png

image.png



illustrates the Geometric distribution, which models the probability of achieving the first success in a series of independent Bernoulli trials.

The probability of success (p) is set to 0.3, meaning there is a 30% chance of success in each trial.

The code begins by calculating the theoretical probabilities for achieving the first success on each trial (k) from 1 to 10 using the formula.

These probabilities are computed with the geom.pmf() function from the scipy.stats library and displayed in a bar chart, where the x-axis represents the trial number and the y-axis shows the corresponding probabilities.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom

p = 0.3  # Probability of success in each trial

# Generate possible outcomes (1 to 10 trials)
k_values = np.arange(1, 11)

# Calculate Geometric probabilities for each trial
geom_probs = geom.pmf(k_values, p)

# Print probabilities
print("Geometric Probabilities:")
for k, prob in zip(k_values, geom_probs):
    print(f"P(X = {k}) = {prob:.4f}")
```

```python
plt.figure(figsize=(8, 5))
plt.bar(k_values, geom_probs, color='skyblue', alpha=0.7, edgecolor='black')
plt.title(f"Geometric Distribution (p = {p})", fontsize=16)
plt.xlabel("Trial Number (k)", fontsize=12)
plt.ylabel("Probability", fontsize=12)
plt.xticks(k_values)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

To demonstrate the practical application of the Geometric distribution, the code simulates 1000 random outcomes using the geom.rvs() function. Each simulated value represents the trial number where the first success occurred. The results are then visualized in a histogram, showing the relative frequencies of each trial number in the simulation. The histogram closely aligns with the theoretical probabilities, illustrating how the Geometric distribution behaves in real-world scenarios. This combination of theoretical and simulated data provides a comprehensive understanding of the Geometric distribution and its applications.

```python
simulated_data = geom.rvs(p, size=1000)  # Generate 1000 trials

# Plot histogram of simulated outcomes
plt.figure(figsize=(8, 5))
plt.hist(simulated_data, bins=np.arange(1, max(simulated_data) + 2), density=True, color='lightgreen', edgecolor='black', alpha=0.7)
plt.title(f"Histogram of Simulated Geometric Outcomes (p = {p})", fontsize=16)
plt.xlabel("Trial Number (k)", fontsize=12)
plt.ylabel("Relative Frequency", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Normal Distribution**

image.png

image.png

image.png

image.png

demonstrates the Normal Distribution, a bell-shaped curve that represents the probability density of data points centered around a mean (μ) with a specific standard deviation (σ).

we define the mean(μ=0) and standard deviation(σ=1).

We then generate 1000 evenly spaced data points between -4 and 4 using np.linspace. The norm.pdf() function calculates the theoretical probability density for each point, creating the smooth bell curve.

This curve is plotted with a blue line, and the mean (μ) is highlighted with a dashed red line to emphasize its central position.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters of the Normal Distribution
mean = 0       # Mean (μ)
std_dev = 1    # Standard Deviation (σ)

# Generate values for the x-axis (data points)
x = np.linspace(-4, 4, 1000)

# Calculate the Probability Density Function (PDF)
pdf = norm.pdf(x, loc=mean, scale=std_dev)

# Plot the Normal Distribution
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, color='blue', label=f'Normal Distribution (μ={mean}, σ={std_dev})')
plt.title("Normal Distribution Curve", fontsize=16)
plt.xlabel("Data Points (x)", fontsize=12)
plt.ylabel("Probability Density", fontsize=12)
plt.grid(alpha=0.4)
plt.axvline(mean, color='red', linestyle='--', label='Mean (μ)')
plt.legend(fontsize=12)
plt.show()
```

the code simulates 1000 random data points following the same Normal distribution using np.random.normal.

 A histogram is plotted to visualize the frequencies of these data points, with 30 bins to categorize the data.

 The histogram represents how the simulated data is distributed, closely aligning with the theoretical Normal distribution curve. The overlay of the theoretical curve on the histogram confirms that the simulated data follows the expected distribution, offering both theoretical and practical insights into how Normal distributions behave.

```python
random_data = np.random.normal(mean, std_dev, 1000)

# Plot histogram of simulated data
plt.figure(figsize=(10, 6))
plt.hist(random_data, bins=30, density=True, color='lightgreen', edgecolor='black', alpha=0.7, label="Simulated Data")
plt.plot(x, pdf, color='blue', label=f'Theoretical Curve (μ={mean}, σ={std_dev})')
plt.title("Histogram of Simulated Normal Data", fontsize=16)
plt.xlabel("Data Points (x)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(alpha=0.4)
plt.legend(fontsize=12)
plt.show()
```

**Continuous Uniform Distribution**

image.png

image.png

image.png



demonstrates the Continuous Uniform Distribution, where all values within a specific range [a,b] are equally likely.

The lower bound (a) is set to 2, and the upper bound (b) is set to 8.

the code generates 1000 evenly spaced values for the x-axis using np.linspace.

The uniform.pdf() function calculates the Probability Density Function (PDF) for these x-values, producing a flat line between a and b since all values in this range have equal probability density.

The PDF is then plotted, with vertical dashed lines marking the lower and upper bounds (a and b). The flat line in the plot visually confirms that the probability is constant within the specified range.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Parameters of the Continuous Uniform Distribution
a = 2    # Lower bound
b = 8    # Upper bound

# Generate values for x-axis
x = np.linspace(a - 1, b + 1, 1000)

# Calculate the PDF
pdf = uniform.pdf(x, loc=a, scale=b-a)

# Plot the PDF
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, color='blue', label=f'Uniform PDF (a={a}, b={b})')
plt.fill_between(x, pdf, where=(x >= a) & (x <= b), color='lightblue', alpha=0.5)
plt.title("Continuous Uniform Distribution - PDF", fontsize=16)
plt.xlabel("Value (x)", fontsize=12)
plt.ylabel("Probability Density", fontsize=12)
plt.axvline(a, color='green', linestyle='--', label='Lower Bound (a)')
plt.axvline(b, color='red', linestyle='--', label='Upper Bound (b)')
plt.grid(alpha=0.4)
plt.legend(fontsize=12)
plt.show()
```

the code uses uniform.rvs() to generate 1000 random values that follow the Uniform distribution within the range[a,b].

A histogram of these random values is created to illustrate their uniform spread. The histogram bars are nearly equal in height, reflecting the equal likelihood of all values in the range.

Vertical dashed lines again mark the bounds (a=2 and b=8), reinforcing the distribution's range. The plots and simulation align closely, providing both theoretical and practical insights into the Continuous Uniform Distribution.

```python
random_data = uniform.rvs(loc=a, scale=b-a, size=1000)

# Plot histogram of the random data
plt.figure(figsize=(10, 6))
plt.hist(random_data, bins=30, density=True, color='lightgreen', edgecolor='black', alpha=0.7, label="Simulated Data")
plt.title("Histogram of Simulated Uniform Data", fontsize=16)
plt.xlabel("Value (x)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.axvline(a, color='green', linestyle='--', label='Lower Bound (a)')
plt.axvline(b, color='red', linestyle='--', label='Upper Bound (b)')
plt.grid(alpha=0.4)
plt.legend(fontsize=12)
plt.show()
```

**Exponential Distribution**

image.png
image.png
image.png

demonstrates the Exponential Distribution, a statistical model often used to represent the time between events that occur at a constant rate.

The rate parameter (λ) is set to 1.5, meaning events occur on average 1.5 times per unit time.

the code generates 1000 evenly spaced values on the x-axis, ranging from 0 to 4, using np.linspace. The expon.pdf() function calculates the probability density for each x-value.

 This creates a decaying curve characteristic of the Exponential Distribution, where shorter time intervals between events (x) have higher probabilities, and longer intervals have progressively lower probabilities.

 The PDF is plotted, showing that the likelihood of an event decreases over time.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Parameters for Exponential Distribution
lambda_param = 1.5  # Rate parameter (λ)

# Generate values for x-axis
x = np.linspace(0, 4, 1000)  # Range of time values (0 to 4)

# Calculate the PDF
pdf = expon.pdf(x, scale=1/lambda_param)

# Plot the PDF
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label=f'Exponential PDF (λ={lambda_param})', color='blue')
plt.fill_between(x, pdf, color='lightblue', alpha=0.5)
plt.title("Exponential Distribution - PDF", fontsize=16)
plt.xlabel("Time between events (x)", fontsize=12)
plt.ylabel("Probability Density", fontsize=12)
plt.grid(alpha=0.4)
plt.legend(fontsize=12)
plt.show()
```

the code simulates 1000 random values using expon.rvs(), which generates data points that follow the Exponential Distribution.

These simulated data points represent the time intervals between consecutive events.

A histogram of these random intervals is then plotted to visualize their frequency. The histogram aligns closely with the theoretical PDF curve, confirming the accuracy of the simulation. Overall, the code effectively demonstrates the mathematical and practical behavior of the Exponential Distribution.

```python
random_data = expon.rvs(scale=1/lambda_param, size=1000)

# Plot histogram of simulated data
plt.figure(figsize=(10, 6))
plt.hist(random_data, bins=30, density=True, color='lightgreen', edgecolor='black', alpha=0.7, label="Simulated Data")
plt.title("Histogram of Simulated Exponential Data", fontsize=16)
plt.xlabel("Time between events (x)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(alpha=0.4)
plt.legend(fontsize=12)
plt.show()
```

**Chi-Square Distribution**

image.png

image.png

image.png



detailed demonstration of the Chi-Square Distribution, a key statistical tool for hypothesis testing and variance analysis.

the number of degrees of freedom (k) is set to 5, which shapes the distribution.

A range of values between 0 and 20 is generated using np.linspace to serve as the x-axis for the theoretical distribution.

The Probability Density Function (PDF) of the Chi-Square Distribution is then computed for these x-values using chi2.pdf. This PDF, which shows how probabilities are distributed, is plotted as a smooth curve. The right-skewed shape of the graph reflects the characteristics of the Chi-Square Distribution for k=5.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Parameters for Chi-Square Distribution
degrees_of_freedom = 5  # Degrees of freedom (k)

# Generate x values for PDF
x = np.linspace(0, 20, 1000)

# Calculate the PDF
pdf = chi2.pdf(x, df=degrees_of_freedom)

# Plot the PDF
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label=f"Chi-Square PDF (df={degrees_of_freedom})", color='blue')
plt.fill_between(x, pdf, color='lightblue', alpha=0.5)
plt.title("Chi-Square Distribution - PDF", fontsize=16)
plt.xlabel("Value", fontsize=12)
plt.ylabel("Probability Density", fontsize=12)
plt.grid(alpha=0.4)
plt.legend(fontsize=12)
plt.show()
```

the code simulates random data representing 1000 observations from a Chi-Square distribution with the same degrees of freedom using chi2.rvs.

These data points are visualized in a histogram, where the height of each bar corresponds to the frequency of values within that range.

The histogram closely resembles the theoretical PDF curve, reinforcing the connection between the mathematical model and real-world simulations.

 This approach helps demonstrate the behavior of the Chi-Square Distribution and its dependence on the degrees of freedom.

```python
random_data = chi2.rvs(df=degrees_of_freedom, size=1000)

# Plot histogram of simulated data
plt.figure(figsize=(10, 6))
plt.hist(random_data, bins=30, density=True, color='lightgreen', edgecolor='black', alpha=0.7, label="Simulated Data")
plt.title("Histogram of Simulated Chi-Square Data", fontsize=16)
plt.xlabel("Value", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(alpha=0.4)
plt.legend(fontsize=12)
plt.show()
```

**Logistic Distribution**
image.png

image.png



demonstrates the Logistic Distribution, highlighting its theoretical properties and practical simulation.

The distribution is parameterized with a location (μ=0) and a scale (s=1).

The location determines the center of the distribution, while the scale controls its spread or steepness.

1000 evenly spaced values in the range −10 to 10 are generated using np.linspace to represent possible outcomes.

The Probability Density Function (PDF) is then calculated for these values using logistic.pdf, which defines the likelihood of different outcomes. The resulting curve is symmetric, bell-shaped, and has heavier tails compared to a normal distribution, indicating higher probabilities for extreme values.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import logistic

# Parameters for the Logistic Distribution
location = 0  # Mean (mu)
scale = 1     # Spread (s)

# Generate x values for PDF
x = np.linspace(-10, 10, 1000)

# Calculate the PDF
pdf = logistic.pdf(x, loc=location, scale=scale)

# Plot the PDF
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label="Logistic PDF", color='blue')
plt.title("Logistic Distribution - PDF", fontsize=16)
plt.xlabel("Value", fontsize=12)
plt.ylabel("Probability Density", fontsize=12)
plt.grid(alpha=0.4)
plt.legend(fontsize=12)
plt.show()
```

To simulate real-world data, 1000 random samples are generated from the logistic distribution using logistic.rvs.

These samples are plotted as a histogram, showing the frequency of the simulated outcomes. The histogram closely matches the PDF curve, demonstrating how theoretical probabilities align with practical simulations.

This visualization helps illustrate the symmetric nature and heavier tails of the logistic distribution, commonly used in logistic regression and growth modeling scenarios.

The code effectively bridges the gap between theory and application by providing a clear, visual understanding of the logistic distribution.

```python
random_data = logistic.rvs(loc=location, scale=scale, size=1000)

# Plot histogram of simulated data
plt.figure(figsize=(10, 6))
plt.hist(random_data, bins=30, density=True, color='lightgreen', edgecolor='black', alpha=0.7, label="Simulated Data")
plt.title("Histogram of Simulated Logistic Data", fontsize=16)
plt.xlabel("Value", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(alpha=0.4)
plt.legend(fontsize=12)
plt.show()
```
