```python
import numpy as np
from scipy import stats

# Define a dataset
data = [10, 20, 20, 30, 100]

# Calculate mean
mean = np.mean(data)

# Calculate median
median = np.median(data)

# Calculate mode
mode = stats.mode(data)

# Display results
print(f"Dataset: {data}")
print(f"Mean: {mean}")
print(f"Median: {median}")
mode = stats.mode(data)
print(f"Mode:{mode[0]}")
```

```python
import numpy as np

# Dataset
data = [5, 10, 15, 20, 25]

# Calculations
range_value = max(data) - min(data)
mean = np.mean(data)
variance = np.var(data)
std_deviation = np.std(data)
mad = np.mean(np.abs(data - mean))
cv = (std_deviation / mean) * 100

# Display results
print(f"Dataset: {data}")
print(f"Range: {range_value}")
print(f"Mean: {mean}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_deviation}")
print(f"Mean Absolute Deviation (MAD): {mad}")
print(f"Coefficient of Variation (CV): {cv:.2f}%")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset
data = [10, 12, 14, 15, 18, 21, 25, 30, 100]

# Calculate Q1, Q3, and IQR
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# Calculate outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

# Print results
print(f"Dataset: {data}")
print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
print(f"Outliers: {outliers}")

# Visualize using a box plot
plt.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title("Box Plot with Outliers")
plt.show()
```
